```python
import torch

from effnetv2_pytorch.effnetv2_model import EffnetV2Model
from torchsummary import summary
```

```python
import timm
```

```python
m = timm.create_model('efficientnetv2_m', pretrained=True)
```

```python
summary(m, (3, 320, 320))
```

```python
summary(model, (3, 320, 320))
```

```python
t = torch.rand((1,3,320,320))
```
